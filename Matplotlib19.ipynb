```python
# # Bar chart
# bar chart rectangular bar or columns are used in bar charts to depict data. they can be used to highlight discrete categories or 
# show changes over time. horizontal bar charts and vertical bar charts are example of different typesa
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [4,3,2,5,6,7,4]
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
names = ['Aman', 'Shreya', 'yash', 'denim']
salaries = [20000, 40000,15000, 50000]

# create bar chart
plt.bar(names, salaries, color = 'skyblue')

# Add labels and titles
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

plt.grid()

# show the plot
plt.show()
```

```python
# # Pie Chart
# Pie charts show data as slices of a circle, each slice representing the percentage of each category. they are useful for 
# showing parts of a whole or contrasting relative sizes.
```

```python
# Pie chart

import matplotlib.pyplot as plt
x = ['Python', 'Java', 'C++', 'Javascript']
y = [40,30,20,10]
plt.pie(y, labels = x, autopct = '%1.1f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['A', 'B', 'C', 'D']
sizes = [15,30,45,10]
colors = ['Gold', 'Lightcoral', 'Lightskyblue', 'Lightgreen']

# create a pie chart
plt.pie(sizes, labels = labels, colors = colors, autopct = '%1.1f%%', startangle = 90)

# add title
plt.title("Pie Chart Example")

# show the plot
plt.show()
```

```python
# Line graph
# line chart data is shown in line charts as a collection of data points connected by lines. they are perfect for displaying trends 
# and patterns along an ongoing axis, like time
```

```python
# Line graph

import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y)
plt.title("Line Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1,2,3,4,5]
y = [2,3,5,7,11]

# create a plot
plt.plot(x, y, marker ='*', linestyle = '--', color = 'b', label = 'Prime Numbers')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Basic line plot")
plt.legend()
plt.grid()

# show the plot
plt.show()
```

```python
# Scatter plot
# Scatter plot individual data points are displayed as dots on a two dimentional grid in scatter plots. they are used to identify
# correlations and illustrate the link between two variables.
```

```python
x = [3,5,8,0,3,7,1,4]
y = [2,4,8,5,3,2,9,4]
plt.scatter(x,y)
plt.show

```

```python
age = [30,15,30,20,13,17,10,40]
Income = [2000, 4000, 80000, 55000, 3300, 21000, 90000, 44000]

# plot the scatter plot
plt.scatter(age, Income, marker = '*', color = 'red')
# plt.grid()
plt.grid(alpha = 0.2)
# plot the label and title
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Scatter Chart Example")

# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# generate random data
data = np.random.randint(1000, size = 500)

# create histogram
plt.hist(data, bins = 30, color = 'green', edgecolor = 'black')

# Add labels and titles
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# dta 
x = np.linspace(0,10,100) #creates an array
y1 = np.sin(x)
y2 = np.cos(x)

# create subplots
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (8,5))

# plot1
ax1.plot(x, y1, color = 'blue')
ax1.set_title('sine wave')

# plot2
ax2.plot(x, y2, color = 'red')
ax2.set_title('Cosine wave')

# Add common labels
fig.suptitle('Multiple Plots Example')
plt.show()
```
