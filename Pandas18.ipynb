```python
# 1 . Pandas Series
# A Pandas series is one dimentional labeled array capable of holding data of any type(integer, string, float, python object etc)
# the axis labels are collectively called indexes
```

```python
# Create first Pandas Series

number_series = pd.Series([2,3,4,5])
print(number_series)
```

```python
number_series = pd.Series([2,3,4,5], index = ['a', 'b', 'c', 'd'], name = 'Number')
print(number_series)
```

```python
# accessing pandas series element
# number_series['c']
number_series[3]
```

```python
# Pandas Dataframe
# Pandas dataframe is a two dimentional data structure with labeled axes(rows and columns)
```

```python
# creating our first dataframe using 2D numpy array

import numpy as np
twod_array = np.random.randint(2,10,(3,4))
first_dataframe = pd.DataFrame(twod_array, columns = ['Akshay', 'Deva', 'Divya', 'Mona'], index = ['a', 'b', 'c'])
first_dataframe
```

```python
# Initialize list of lists

import numpy as np
import pandas as pd
list_of_lists = [['Ashu', 10], ['raghav', 20], ['shila', 30]]

# create pandas dataframe
df = pd.DataFrame(list_of_lists, columns = ['Name', 'Age'])

# print dataframe
df
```

```python
# create dataframe using dictionary of lists(column wise) --- dictionary of lists must have equal-length lists:
import pandas as pd

# store the data in the dictionary
data = {
    'Name' : ['Alice', 'Bob', 'Charlie'],
    'Age' : [25,30,35],
    'City' : ['New york', 'London', 'Paris']
}

# create dataframe
df = pd.DataFrame(data)

# print dataframe
print(df)
```
