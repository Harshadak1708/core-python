```python
# Importing Seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Dataset
# Seaborn comes with built-in datasets that you can use for practise. Let's load the "tips" dataset
```

```python
# Load the tips dataset

tips = sns.load_dataset("tips")
print(tips.head())
```

```python
tips
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips)
plt.title("total bill Vs tips")
plt.grid()
plt.show()
```

```python
# Line plot
sns.lineplot(x = "total_bill", y = "tip", data = tips)
plt.title("total bill Vs tips")
plt.grid()
plt.show()
```

```python
# Bar plot
sns.barplot(x = "day", y = "total_bill", data = tips)
plt.title("Bar plot day Vs total_bill")
plt.show()
```

```python
sns.histplot(tips['total_bill'], kde = True)
plt.title("Histplot of totalbill")
plt.show()
```

```python
# Box plot
# a box plot (also called box and whisker plot) is a graph that shows the spread and distribution of a dataset using 5 key summary values.
```

```python
sns.boxplot(x = "day", y = "total_bill", data = tips)
plt.title("total bill by day(box plot)")
plt.show
```

```python
# Pair Plot
# A pairplot is a powerful visualization tool that shows pairwise relationships between multiple columns in a dataset â€” especially
# useful for exploring numerical data.

# A pairplot creates:

# Scatter plots between every pair of numerical columns

# Histograms (or KDEs) on the diagonal for individual columns
```

```python
sns.pairplot(tips, hue = "sex")

plt.show
```

```python
# Heatmap
# A heatmap is a data visualization technique that uses color to represent the magnitude of values in a matrix or 2D data.
# Or you can say a heatmap is like a colorful table â€” instead of just showing numbers, it uses colors to make patterns and trends easy to see.
```

```python
# Heat map
# selected the numeric column
selected_columns = ['total_bill', 'tip', 'size']

corr = tips[selected_columns].corr()

# plot heatmap
sns.heatmap(corr, annot = True,  cmap = 'YlGnBu')
plt.title("correlation heatmap(Selected columns)")
plt.show()
```

```python
# Violin Plot
# Violin Plot is a method to visualize the distribution of numerical data of different variables. It is quite similar to Box Plot but with
# a rotated plot on each side, giving more information about the density estimate on the y-axis.

# A violin plot consists of four components.

# A white Centered Dot at the middle of the graph - The white dot point at the middle is the median of the distribution.
# A thin gray bar inside the plot - The bar in the plot represents the Quartile range of the distribution.
# A long thin line coming outside from the bar - The thin line represents the rest of the distribution which is calculated by the formulae Q1-1.5
# IQR for the lower range and Q3+1.5 IQR for the upper range. The point lying beyond this line are considered as outliers.
# A line boundary separating the plot- A KDE plot is used for defining the boundary of the violin plot it represents the distribution of data points.
```

```python
# Create a violin plot
sns.violinplot(x='day', y='tip', data=tips)

# Add title and show the plot
plt.title('Violin Plot of Tips by Day')
plt.show()
```

```python
# What this shows:

# Distribution of tips for each day (Thu, Fri, Sat, Sun).

# The width of the violin indicates the density of data points at different values.
```
